## 01. Tips data set

```python
sns.set_style('whitegrid')

tips = sns.load_dataset('tips') # 인터넷 연결 필요
tips
```

```python
import matplotlib as mpl
mpl.rc('font', family='Malgun Gothic') ## 맑은 고딕 
```

```python
# 요일 별 total_bill
plt.figure(figsize = (8, 6))
sns.boxplot(x = "day", y = "total_bill", data = tips)
plt.title("요일 별 total_bill")
plt.show()
```

```python
# (실습) 요일별 인원 수
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="size", data=tips)

plt.title("요일 별 인원 수")
plt.show()
```

```python
# 요일별 팁
plt.figure(figsize=(8,6))
sns.boxplot(x = "day", y="tip", data=tips)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, size=7)
plt.show()
```

## 02. flights

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fg = sns.load_dataset("flights")
fg.head(5)
fg
```

```python
fgp = fg.pivot("month", "year", "passengers")
fgp
```

```python
## sns.boxplot
## sns.lmplot
## sns.heatmap
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(fgp)
plt.show()
```

```python
## 숫자 표시 (annot = True)
plt.figure(figsize=(10,8))
sns.heatmap(fgp, annot=True)
plt.show()
```

```python
## 정수로 표시 (fmt = 'd')
plt.figure(figsize=(10,8))
sns.heatmap(fgp, annot=True, fmt='d')
plt.show()
```
