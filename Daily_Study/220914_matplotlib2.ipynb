### 07. Super Title

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,5))

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 1행 2열, 첫번째
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")


#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 1행 2열, 두번째
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP") # super title
plt.show()
```

(실습) 2행, 2열의 그래프를 표시해보자. 2행은 산점도 등의 다른 그래프 종류로 표시

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,5))

# plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 1행 2열, 첫번째
plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title("Plot1")

# plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 1행 2열, 두번째
plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.title("Plot2")

# plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 1행 2열, 두번째
plt.subplot(2, 2, 3)
plt.plot(x, y, 'o')
plt.title("Plot3")

# plot 4:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 1행 2열, 두번째
plt.subplot(2, 2, 4)
plt.plot(x,y, 'ro')
plt.title("Plot4")


plt.suptitle("MY SHOP")
plt.tight_layout()
plt.show()
```

### 08. Scatter plot

- 직교 좌표계를 활용하여 좌표상의 점들을 표시.
- 두개 변수 간의 관계를 나타낼 수 있다.

```python
x = np.array([51,71,18,17,21])
y = np.array([99,86,87,88,111])

plt.scatter(x, y)
plt.show()
```

```python
x = np.array([51,71,18,17,21,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y, color='hotpink')

x = np.array([511,711,118,17,21,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y, color='#88c999')
plt.show()
```

- Numpy 배열 형태로 만들어주면 속도가 훨씬 빠르다.
- Numpy 배열은 C언어, C++로 만들어졌다.

```python
x = np.random.randint(10, size=10) # 0~9까지 랜덤으로 정수 난수 10개 생성
y = np.random.randint(80,100, size=10) # 80~99까지 랜덤으로 정수 난수 10개 생성

print(x, y)
print("length x:", len(x))
print("length y:", len(y))
```

```python
# x = np.random.randint(10, size=10)
# y = np.random.randint(80,100, size=10)
# print(x, y)
# print( len(x), len(y) )

# 색에 숫자로 맵핑
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 100])
# 'viridis', 'plasma', 'inferno', 'magma', 'cividis'

plt.scatter(x, y, c=colors, cmap='magma', s=150)
plt.colorbar() # colormap 을 포함할 수 있다.

plt.show()
```

### 09. Alpha (투명도)

```python
x = np.random.randint(10, size=10)
y = np.random.randint(80,100, size=10)

sizes = np.random.randint(20,800, size=10)
colors = np.random.randint(0,100, size=10)

plt.scatter(x, y, s=sizes, alpha=0.5)
plt.show()
```

- `plt.plot()` : 산점도, 선
- `plt.scatter()` : 산점도
- `plt.bar()` : 막대그래프

```python
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')
plt.show()
```

### 12. Histogram

- 하나의 변수의 분포를 표현해 줄 때 사용
- 연속형 값을 표시할 떄 사용
- 가로축이 계급, 세로축이 도수(구간의 값의 개수)를 의미

```python
# 정규 분포를 따르는 값. 100개 생성
x = np.random.normal(loc=170, scale=10, size=100) # 평균, 표준편차, 개수

print(x)
plt.hist(x)
plt.show()
```

### 13. pie chart (원 그래프)

```python
y = np.array([30,20,35,10, 5])
plt.figure(figsize=(10,10))
plt.pie(y)
plt.show()
```

```python
plt.figure(figsize=(10,10))

y = np.array([30,20,35,10, 5])
addlbl = ['사과', '바나나', '딸기', '수박', '오렌지']
plt.pie(y, labels=addlbl)
plt.show()
```

```python
y = np.array([30,20,35,10, 5])
addlbl = ['사과', '바나나', '딸기', '수박', '오렌지']
plt.pie(y, labels = addlbl, startangle = 90)
plt.show()
```

```python
y = np.array([30,20,35,10, 5])
addlbl = ['사과', '바나나', '딸기', '수박', '오렌지']
myexplode = [0.2, 0, 0, 0, 0.3]

plt.pie(y, labels = addlbl, explode = myexplode)
plt.show()
```

```python
y = np.array([30,20,35,10, 5])
addlbl = ['사과', '바나나', '딸기', '수박', '오렌지']
myexplode = [0.2, 0, 0, 0, 0.3]

plt.pie(y, labels = addlbl, explode = myexplode)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))

y = np.array([30,20,35,10, 5])
addlbl = ['사과', '바나나', '딸기', '수박', '오렌지']
myexplode = [0.2, 0, 0, 0, 0.3]

plt.pie(y, labels = addlbl)
plt.legend(title = "5가지 과일")
plt.show()
```
