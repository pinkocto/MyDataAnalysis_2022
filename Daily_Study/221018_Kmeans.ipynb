# 군집(Clustering) - Kmeans

```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
```

```python
# 데이터 만들기(2차원 데이터)
X, y = make_moons(n_samples=200, noise=0.05, random_state=0) 
print(X.shape, y.shape)
```

```python
import matplotlib.pyplot as plt

# 클러스터 할당과 클러스터 중심을 표시한다.
# 특성 1, 특성 2
X1 = X[: ,0]
X2 = X[: ,1]
plt.scatter(X1, X2, 
            c=y, s=60, edgecolors='k')
```

```python
# 두 개의 클러스터로 데이터에 KMeans 알고리즘 적용
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_pred = kmeans.predict(X)
y_pred
```

```python
import matplotlib.pyplot as plt

# 클러스터 할당과 클러스터 중심을 표시한다.
X1 = X[: ,0]
X2 = X[: ,1]
plt.scatter(X1, X2, c=y_pred, s=60, edgecolors='k')
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt


# 인위적으로 2차원 데이터 생성
X, y = make_blobs(random_state=1)

# 데이터 확인
print(X.shape)
print(y.shape)
print(np.unique(y))  # y : 0,1,2를 갖는 값.

# 군집 모델 만들기 (그룹이 3개)
model = KMeans(n_clusters=3)
model.fit(X)

# 레이블 확인
print("클러스터 레이블:\n{}".format(model.labels_))

print("예측값")
# 군집화 시키기
print(model.predict(X))

X1 = X[:, 0] # 첫번째 열
X2 = X[:, 1] # 두번째 열

plt.scatter(X1, X2, c=y)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
```

```python
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)

# 평균 0, 분산 1이 되도록 데이터 스케일 조정
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
dbscan = DBSCAN()  # eps = 0.5, min_samples=5
clusters = dbscan.fit_predict(X_scaled)
clusters
```

```python
import matplotlib.pyplot as plt

# 원래 데이터 그래프
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)   # 2행 1열에 첫번째 
f1 = X_scaled[:, 0]  # 특성 첫번째 선택
f2 = X_scaled[:, 1]  # 특성 두번째 선택
plt.scatter(f1, f2, c=y)  # 원본 샘플 데이터 
plt.xlabel("특성 0")
plt.ylabel("특성 1")
plt.title("ORI DATA")

# DBSCAN 적용한 그래프
plt.subplot(1,2,2)   # 2행 1열에 두번째 
f1 = X_scaled[:, 0]  # 특성 첫번째 선택
f2 = X_scaled[:, 1]  # 특성 두번째 선택
plt.scatter(f1, f2, c=clusters) # DBSCAN 적용한 샘플 데이터
plt.xlabel("특성 0")
plt.ylabel("특성 1")
plt.title("DBSCAN")

plt.show()
```

### 3-4 실습 eps에 따른 그룹은 어떻게 변경되는지 확인해 보자.(eps = 0.1, 0.5, 1, 2, 4)

```python
dbscan = DBSCAN(eps=0.1)  # eps = 0.5, min_samples=5
clusters = dbscan.fit_predict(X_scaled)
clusters

# DBSCAN 적용한 그래프
plt.subplot(1,2,2)   # 2행 1열에 두번째 
f1 = X_scaled[:, 0]  # 특성 첫번째 선택
f2 = X_scaled[:, 1]  # 특성 두번째 선택
plt.scatter(f1, f2, c=clusters) # DBSCAN 적용한 샘플 데이터
plt.xlabel("특성 0")
plt.ylabel("특성 1")
plt.title("DBSCAN")

plt.show()
```
